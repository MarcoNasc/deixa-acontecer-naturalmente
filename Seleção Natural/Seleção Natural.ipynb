# Seleção Natural

## Neste trabalho, iremos analisar o comportamento de uma população de coelhos em relação a uma combinação de diferentes parâmetros, por meio de um simulador disponibilizado pela University of Colorado Boulder em seu [site](https://phet.colorado.edu/en/simulation/natural-selection), onde encontramos diversas outros simuladores interativos com fins educacionais.


![title](images/site.png)

## Neste simulador, temos a disponibilidade três diferentes fenótipos à disposição, assim como dois fatores de selação e também dois ambientes distintos. Eles são: cor da pelagem, tamanho da cauda e tamanho dos dentes; predadores e comida; ártico e equador; respectivamente.

## Serão feitos uma séries de experimentos, alternando entre os diversos parâmetros a fim de descobrir os efeitos produzidos na população de coelhos após tal mudança, em seguida serão criadas hipóteses explicando esse comportamente e por fim mais alguns experimentos irão de ser realizados com o objetivo de confirmar as hipóteses.

### Experimento 01: Pelagem escura


```python
% matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
dados = [2, 4, 6]
iterations = [1, 2, 3]

mean = sum(dados)/len(dados)


fig, ax = plt.subplots()
ax.plot(iterations, dados, label = "Ex01", color='r')
ax.set_xlabel('Iterações')
ax.set_ylabel('População')
ax.set_title("Experiência 01")
ax.plot(iterations, [mean for i in iterations], label='Mean', linestyle='--')
```

```python
print(f'Média: {np.mean(dados)}')
print(f'Desvio Padrão: {np.std(dados)}')
```
